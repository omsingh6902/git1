## Expand the topic-topic data to split the rows according to multiple combinations of streams

```python
import pandas as pd

def expand_stream_combinations(row):
    """
    For a given row, generate all possible combinations of topic_stream and prerequisite_stream.
    """
    # Split the streams into lists
    topic_streams = row['topic_stream'].split(', ')
    prerequisite_streams = row['prerequisite_stream'].split(', ')
    
    # Generate all combinations
    combinations = []
    for t_stream in topic_streams:
        for p_stream in prerequisite_streams:
            combinations.append({
                'topic_class': row['topic_class'],
                'topic': row['topic'],
                'prerequisite': row['prerequisite'],
                'prerequisite_class': row['prerequisite_class'],
                'topic_stream': t_stream,
                'prerequisite_stream': p_stream
            })
    return combinations

def process_csv(file_path):
    """
    Reads a CSV file and expands it into all possible combinations of topic_stream and prerequisite_stream.
    """
    # Load the CSV file
    data = pd.read_csv(file_path)

    # Apply the function to each row and explode the dataframe to accommodate all combinations
    expanded_rows = data.apply(expand_stream_combinations, axis=1).explode().reset_index(drop=True)

    # Convert the list of dictionaries into a DataFrame
    expanded_df = pd.DataFrame(expanded_rows.tolist())

    return expanded_df

# Define the path to your CSV file
file_path = '/Users/omsingh/Downloads/new_KG_topic_links_ADPL - biology-final.csv'  # Replace this with the actual path to your CSV file

# Process the CSV file
expanded_df = process_csv(file_path)

# Optionally, save the expanded dataframe to a new CSV file
expanded_df.to_csv('/Users/omsingh/Downloads/expanded_data_bio.csv', index=False)

# Display the first few rows of the expanded dataframe
print(expanded_df.head())
```

#### Matching each subtopic to its stream while taking class into consideration

This is the most important step because GPT API may have hallucinated while generating the links in a manner such that for a particular topic, its either creating a totally new subtopic name or its using a topic name which is actually present in some other class but seems logically correct. 

Below is a script which matches each topic and prererequisite to its stream while taking class into consideration and adds it to two new **columns topic_stream** and **prerequisite_stream**

```python
import pandas as pd
import ast

# Load the datasets
kg_df = pd.read_csv('/Users/omsingh/Downloads/KG_Subtopic_Links_ADPL - final_maths_all.csv').dropna(axis=0)
adpl_df = pd.read_csv('/Users/omsingh/Downloads/final_subtopic_links/ADPL_New_Reference_Subtopic_View - math-all.csv')

# Parse the Subtopics in the ADPL DataFrame
adpl_df['Subtopics'] = adpl_df['Subtopics'].apply(ast.literal_eval)

# Flatten the ADPL DataFrame for easy lookup
flat_list = []
for index, row in adpl_df.iterrows():
    for subtopic in row['Subtopics']:
        flat_list.append({'Stream': row['Stream'], 'Class': row['Class'], 'Topic': row['Topic'], 'Subtopic': subtopic})
adpl_flat_df = pd.DataFrame(flat_list)

# Initialize new columns in KG DataFrame
kg_df['subtopic_exists'] = False
kg_df['prerequisite_exists'] = False

# Update the KG DataFrame based on the ADPL DataFrame
for index, row in kg_df.iterrows():
    # Check for subtopic existence
    if any((adpl_flat_df['Stream'] == row['subtopic_stream']) &
           (adpl_flat_df['Class'] == row['topic_class']) &
           (adpl_flat_df['Topic'] == row['topic']) &
           (adpl_flat_df['Subtopic'].str.lower() == row['subtopic'].lower())):
        kg_df.at[index, 'subtopic_exists'] = True
    
    # Check for prerequisite existence
    if any((adpl_flat_df['Stream'] == row['prerequisite_stream']) &
           (adpl_flat_df['Class'] == row['prerequisite_class']) &
           (adpl_flat_df['Topic'] == row['prerequisite_topic']) &
           (adpl_flat_df['Subtopic'].str.lower() == row['prerequisite'].lower())):
        kg_df.at[index, 'prerequisite_exists'] = True

# Save the updated KG DataFrame
kg_df.to_csv('/Users/omsingh/Downloads/updated_KG_Subtopic_Links_math1.csv', index=False)
```

```python
# Remove all rows where either 'subtopic_exists' or 'prerequisite_exists' is False
filtered_df = kg_df[(kg_df['subtopic_exists'] == True) & (kg_df['prerequisite_exists'] == True)]

# Save the filtered dataframe to a new CSV file
filtered_file_path = '/Users/omsingh/Downloads/filtered_KG_Subtopic_Links_math1.csv'
filtered_df.to_csv(filtered_file_path, index=False)

filtered_file_path
```
